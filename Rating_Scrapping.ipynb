```python
#Importing the libraries

import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
```

```python
url = "https://www.flipkart.com/"
driver.get(url)
```

```python
#Closing login pop-up
close=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
close.click()
```

##### Laptop

```python
#Entering Laptop in search bar

driver.find_element_by_xpath("//input[@class='_3704LK']").send_keys("laptop")
time.sleep(2)
```

```python
#Clicking on search button
search=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search.click()
```

```python
#appending urls of product
urls=[]
for i in range(10):
    for j in driver.find_elements_by_xpath("//div[@class='_2kHMtA']/a"):
        urls.append(j.get_attribute("href"))
    try:
        driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]").click() #Clicking next button
    except:
        driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
    time.sleep(4)
```

```python
rating=[]
review=[]
for k in urls:
    driver.get(k)
    time.sleep(3)
    try:
        driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']").click() # click on see all reviews
        time.sleep(2)
        # fetching first pages reviews
        for a in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
            review.append(a.text)
        for b in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
            rating.append(b.text)
        time.sleep(1)
        #clicking on second page for reviews
        driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        time.sleep(2)
        # running loop till the end of pages for all reviews
        while True:
            time.sleep(2)
            for i in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
                review.append(i.text)
            for j in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
                rating.append(j.text)
            try:
                driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]/span").click()
                time.sleep(1)
            except:
                break
    except:
        print("NaN")
```

```python
df1=pd.DataFrame({})
df1['Full_Reviews']=review
df1['Ratings']=rating
df1
```

```python
# saving as csv file
df1.to_csv("laptop.csv")
```
